```python
# break and continue
```

```python
# break effectively stops the loop
```

```python
for i in range(2,10):
  if i*5 > 20:
    break
  print(i)
```

```python
for i in range(2,10):
  if i*5 == 20:
    continue
  print(i)
```

```python
runners = ["Yagyansh Bagri", "Arun Krisghnamurthy", "Ansh Bhatnagar"]
distances = ['5k', '10k', 'Marathon']
```

```python
for runner in runners:
  for distance in distances:
    print(runner, ":", distance)
```

```python
sum = 100

for x in range(5):
  sum += x*x
  print(sum)
```

```python
list = [22, 55, 88]
sum = 50

for x in list:
  sum += x
  print(sum)
```
