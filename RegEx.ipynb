```python
# Key differences between match and search
# re.match() checks for matches only at the start of the string,
# while re.search() looks for matches anywhere in the string.
```

```python
all_games = "1920 Antwerp Olympics, 1924 Paris Olympics, 1928 Amsterdam Olympics"
```

```python
find_all_olympics = re.findall("Olympics", all_games)
```

```python
find_all_olympics
```

```python
type(find_all_olympics)
```

```python
len(find_all_olympics)
```

```python
find_all_olympics_iter = re.finditer("Olympics", all_games)
```

```python
find_all_olympics_iter
```

```python
for olympics in find_all_olympics_iter:
  match_text = olympics.group()
  start_index = olympics.start()
  end_index = olympics.end()
  span = olympics.span()
  print(f"Match: {match_text}, Start Index: {start_index}, End Index: {end_index}, Span: {span}")
```

```python
text = "ab a abbb abbbb a a abb"
```

```python
re.findall(r"ab*", text) # Matches a followed by zero or more b's
```

```python
re.findall(r"ab+", text) # Matches a followed by one or more b's
```

```python
re.findall(r"ab?", text) # Matches a followed by zero or one b
```
