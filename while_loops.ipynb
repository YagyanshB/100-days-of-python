```python
# common with while loops is defining a variable outside
```

```python
x = 4

while x <= 10:
  print(x)
  x += 1
```

```python
for x in range(1,100):
  if x % x == 0 and x % 2 != 0:
    print(x)
```

```python
x = 1

sum = 0

while x < 10:
  sum += x
  x += 1

print(sum)
```

```python
# infinite loops
```

```python
x = 1

while x < 10:
  if x == 9:
    break
  print(x)
  x += 1
```

```python
x = 1

while x < 10:
  if x == 9:
    x += 1
    continue
  print(x)
  x += 1
```

```python
# while loops with input

while True:
  x = input("Enter a number: ")
  if x == 'z':
    break
  print(int(x) * int(x))
```

```python
x = input("Enter a positive number less than 20: ")

while int(x) < 0 or int(x) >= 20:
  print("Entered wrong number")
  x = input("Enter a positive number less than 20: ")

print(int(x) * int(x))
```

```python
from typing_extensions import OrderedDict
# while loops with lists

imaginary_numbers = [8921, 1292, 2809, 2409, 1409]

sum = 0
even_sum = 0
odd_sum = 0

x = 0

while x < len(imaginary_numbers):
  if imaginary_numbers[x] % 2 == 0:
    even_sum += imaginary_numbers[x]
  else:
    odd_sum += imaginary_numbers[x]
  x += 1

print('The sum of even numbers is', ":", even_sum)
print('The sum of odd numbers is', ":", odd_sum)
```

```python
# multiple while loops

x = 1

while x <= 8:
  y = 1
  while y <= 3:
    result = x*y
    print(x, '*', y, '=', result)
    y += 1
  x += 1
```
