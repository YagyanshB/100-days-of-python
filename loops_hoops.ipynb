```python
# Example 1 with a basic if else statement
```

```python
batting_average = 0.33
```

```python
if batting_average > 0.3:
  print("Good Batter")

elif batting_average >= 0.25:
  print("Decent Batter")

else:
  print("Needs to Improve")
```

```python
batting_average = float(input("Please enter your batting Average: "))

if batting_average > 0.3:
  print("Good Batter")

elif batting_average >= 0.25:
  print("Decent Batter")

else:
  print("Needs to Improve")
```

```python
# Example 2 with Multiple Conditions
```

```python
hits = int(input("Please enter your total number of hits: "))
home_runs = int(input("Please enter your total number of home runs: "))

if hits > 150 and home_runs > 30:
  print("Great all round player")

elif hits > 150:
  print("Great contact hitter")

elif home_runs > 25:
  print("Great power hitter")

else:
  print("Average Player")
```

```python
# Example 3 with True conditions
```

```python
fast_runner = True

outs = 2
balls = 3
strikes = 2
```

```python
if fast_runner or (outs == 2 and balls == 3 and strikes == 2):
  print("Can Steal Base")

else:
  print("Can't Steal Base")
```
